## Convert dataframe
Convert the cleaned wide dataframe with each row corresponding to each participant to a long dataframe with each row corresponding to each participant-painting pair

### Logistic Regression

```python
# Add a column of ones for linear regression intercept
# Refer to the external resource found at here
# https://stackoverflow.com/questions/29517072/add-column-to-dataframe-with-constant-value
df_logistic['Coefficient'] = 1
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_logistic, test_size=0.3, random_state=5, shuffle=True)
```

```python
# Perform Logistic Regression
# Refer to the external resource found at here
# https://www.geeksforgeeks.org/logistic-regression-using-statsmodels/
import statsmodels.api as sm
X_train = train[['Coefficient', 'Colours_Like', 'Brushstrokes_Like', 'Structures_Like', 'Details_Like', 'Condition_Control', 'Condition_Treatment1',
            'Condition_Treatment2', 'Monthly_Art_Events', 'Monthly_Hours_Digital_Arts', 'Artistic_Specialty_Self_Rating', 'Painting_GroundTruth_Human',
                 'Painting_GroundTruth_AI']]
Y_train = train[['Identification_Correctness']]

#fit linear regression model
model = sm.Logit(Y_train, X_train).fit()

#view model summary
print(model.summary())
```

```python
X_test = test[['Coefficient', 'Colours_Like', 'Brushstrokes_Like', 'Structures_Like', 'Details_Like', 'Condition_Control', 'Condition_Treatment1',
            'Condition_Treatment2', 'Monthly_Art_Events', 'Monthly_Hours_Digital_Arts', 'Artistic_Specialty_Self_Rating', 'Painting_GroundTruth_Human',
                 'Painting_GroundTruth_AI']]
Y_test = test[['Identification_Correctness']]

Y_hat = model.predict(X_test)
Y_pred = list(map(round, Y_hat))
```

```python
# Compute the accuracy score of the logistic regression
# Refer to the external resource found at here
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```
